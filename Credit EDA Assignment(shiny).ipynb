# Merging all the secondary Documents Column into 1 and droping the rest

```python
# Merging all the secondary Documents Column into 1 and droping the rest
app['add_docs'] = app.FLAG_DOCUMENT_2+app.FLAG_DOCUMENT_3+app.FLAG_DOCUMENT_4+app.FLAG_DOCUMENT_5+app.FLAG_DOCUMENT_6+app.FLAG_DOCUMENT_7+app.FLAG_DOCUMENT_8+app.FLAG_DOCUMENT_9+app.FLAG_DOCUMENT_10+app.FLAG_DOCUMENT_11+app.FLAG_DOCUMENT_12+app.FLAG_DOCUMENT_13+app.FLAG_DOCUMENT_14+app.FLAG_DOCUMENT_15+app.FLAG_DOCUMENT_16+app.FLAG_DOCUMENT_17+app.FLAG_DOCUMENT_18+app.FLAG_DOCUMENT_19+app.FLAG_DOCUMENT_20+app.FLAG_DOCUMENT_21
```

```python
# checking values in add_docs
app.add_docs.value_counts()
```

```python
# dropping flag_documents columns
app.drop(app.loc[:,'FLAG_DOCUMENT_2':'FLAG_DOCUMENT_21'].columns, axis= 1, inplace= True)
```

```python
# Checking for null or missing values
temp = (100*app.isna().sum()/app.shape[0]).reset_index().rename(columns={0:"missing"})
temp
```

```python
# dropping NAME_TYPE_SUITE column
app.drop('NAME_TYPE_SUITE', axis= 1, inplace= True)
```

```python
# dropping CODE_GENDER column
app.drop('CODE_GENDER', axis= 1, inplace= True)
```

```python
## Fixing the Columns Data
```

```python
# checking value count if customer own a car or not
app.FLAG_OWN_CAR.value_counts()
```

```python
#Changing FLAG_OWN_CAR values from N/Y to 0/1
app.FLAG_OWN_CAR = app.FLAG_OWN_CAR.apply(lambda x: 0 if x=='N' else 1)
```

```python
# checking value count if customer own a property or not
app.FLAG_OWN_REALTY.value_counts()
```

```python
#Changing FLAG_OWN_REALTY values from N/Y to 0/1
app.FLAG_OWN_REALTY = app.FLAG_OWN_REALTY.apply(lambda x: 0 if x=='N' else 1)
```

```python
# checking values DAYS_BIRTH
app.DAYS_BIRTH
```

```python
#Converting Age days into years and removing the -ve sign
app['age'] = app.DAYS_BIRTH.apply(lambda x: int((x/365)-((x/365)*2)))
app.age
```

```python
# dropping REGION_POPULATION_RELATIVE column
app.drop('REGION_POPULATION_RELATIVE', axis= 1, inplace= True)
```

```python
# checking values DAYS_EMPLOYED
app.DAYS_EMPLOYED
```

```python
#Converting Employment days into years and removing the -ve sign
app['employed'] = app.DAYS_EMPLOYED.apply(lambda x: int((x/365)-((x/365)*2)))
app.employed
```

```python
#Converting last registration days into years and removing the -ve sign
app['registration'] = app.DAYS_REGISTRATION.apply(lambda x: int((x/365)-((x/365)*2)))
app.registration
```

```python
#Converting last name update days into years and removing the -ve sign
app['ID'] = app.DAYS_ID_PUBLISH.apply(lambda x: int((x/365)-((x/365)*2)))
app.ID
```

```python
# dropping less useful columns
app.drop(app.loc[:,'REGION_RATING_CLIENT_W_CITY':'HOUR_APPR_PROCESS_START'].columns, axis= 1, inplace= True)
```

```python
# checking null values in previous application dataframe
100*prev.isna().sum()/prev.shape[0]
```

```python
# dropping columns with more than 90% null values
temp = (100*prev.isna().sum()/prev.shape[0]).reset_index().rename(columns={'index':'cols',0:'missing'})
drop_cols = list(temp.cols[temp['missing']>90])
prev.drop(drop_cols, axis= 1, inplace= True)
```
